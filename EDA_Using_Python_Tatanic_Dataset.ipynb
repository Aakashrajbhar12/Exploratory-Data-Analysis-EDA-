### Missing Value Imputation

```python
# Heatmap to see Null Value proportion
cols = data1.columns 
colours = ['#000000', '#ffff00'] # specify the colours - yellow is missing. black is not missing.
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colours))

#Inference: As it can be seen, 'Age' & 'Cabin' Coln has missing value.
```

```python
#To get count the no of misssing value
data1.isnull().sum()

#Inference:'Age': 177 NA Values, 'Cabin': 687 NA values, 'Embarked': 2 NA values
```

```python
#Mean imputation For Age Coln -- Taking Mean of Age Coln and filling NA values with mean Value
mean = round(data1.Age.mean())   #Using Round fuction bcoz age cannot be decimal
print(mean)
```

```python
#To fill NA value of 'Age' with its mean
data1['Age'] = data1['Age'].fillna(mean)

#To check 'Age' is filled or not
data1.info()

#inference: 'Age' is now filled
```

### Dealing with 'Cabin' column

This coln has almost 80% missing value, So you cant derived appropriate result using this coln,But before Dropping a Coln / feature you have to take suggestion from domain experts

As im practicing i'll do both value imputation of 'Cabin' and also drop 'Cabin' column

```python
#Imputation of value for 'Cabin' coln

#Checking missing value count
data1.Cabin.isnull().sum()
```

```python
#Taking 'Cabin' coln into another dataframe object to impute
objcol=data1[['Cabin']]
```

```python
#Missing value imputation for categorical value of 'Cabin'
objcol=objcol.fillna(objcol.mode().iloc[0])

#Checking missing value count after imputation --it should be Zero
objcol.isnull().sum()
```

```python
#To Drop the cabin coln and create a different data set

#Through this you can drop a column if most value are missing
data2=data1.drop('Cabin', axis=1)
```

```python
#Join the data set with imputed object of cabin i.e: objcol

data2=pd.concat([data2,objcol],axis=1)
data2.head()
```

```python
#Embarked Column -- 2 missing value 

#Getting count of NA value
data2.Embarked.isnull().sum()
```

```python
#Taking 'Cabin' coln into another dataframe object to impute
objcol1=data2[['Embarked']]
```

```python
#Missing value imputation for categorical value of 'Embarked'
objcol1 = objcol1.fillna(objcol1.mode().iloc[0])

#Checking missing value count after imputation --it should be Zero
objcol1.isnull().sum()
```

```python
#To Drop the cabin coln and create a different data set

#Through this you can drop a column if most value are missing
data3=data2.drop('Embarked', axis=1)
```

```python
#Join the data set with imputed object of cabin i.e: objcol1

data3=pd.concat([data3,objcol1],axis=1)
data3.head()
```

```python
#Final Check for missing Values
data3.info()

#Inferences as you can see there are now no missing value all column have non-null in it
```
