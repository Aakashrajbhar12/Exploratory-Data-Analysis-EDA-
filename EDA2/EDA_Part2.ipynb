#### Isolation Forest to Detect Outliers

```python
#Importing the Required Library
from sklearn.ensemble import IsolationForest
```

```python
#Loading the Dataset
data = pd.read_csv("C:/Users/Akaash/Downloads/iris.csv",index_col=0)
#One hot Encoding
data_encoded=pd.get_dummies(data)
```

```python
# training the model 
clf = IsolationForest(random_state=10,contamination=.01) #contamination = 0.1 so telling there are 10% outliers
clf.fit(data_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

Inference: -1 are the outliers

```python
#Checking the Dataset for value
data_encoded
```

```python
#Let us add a new data point which is outlier
data_encoded.loc[150]=[20,40,30,50,1,0,0]
data_encoded.tail()
```

Inference: The 150 records value are changed now rerun  the Model to check if it predict 150th records as outliers

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

Inference: It predicted the 150th records as outliers

```python
#Anomaly Scores
data_encoded['scores']=clf.decision_function(data_encoded)
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:7])
data_encoded
```

Inference: Anomaly Score -1 are Outliers so 150th record is an Outlier

```python
#Print the outlier data points
data_encoded[data_encoded['anomaly']==-1]
```

Inference: Checking Anomaly Score -1 are Outliers so, 107th & 150th record is an Outlier
