#### Isolation Forest to Detect Outliers

```python
#Importing the Required Library
import numpy as np
from sklearn.ensemble import IsolationForest
```

```python
#Loading the Dataset
data = pd.read_csv("C:/Users/Akaash/Downloads/Titanic.csv")
#One hot Encoding
data_encoded=pd.get_dummies(data)
```

```python
# training the model 
clf = IsolationForest(random_state=10,contamination=.01) #contamination = 0.1 so telling there are 10% outliers
clf.fit(data_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

Inference: -1 are the outliers

```python
#List of Outliers Datapoint
list(np.where(y_pred_outliers==-1))
```

Inference: Give the List of where there are Outliers

```python
data_encoded.loc[1490]
```

Inference: This is one of the Datapoint which is an outlier now will assign the some value to another Datapoint and Check the Algorithm for that Datapoint

```python
#Let us add a new data point which is outlier
data_encoded.loc[1]=[1,0,0,0,0,1,0,1,0,1]
data_encoded.head()
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.02)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

Inference: Now See Datapoint of Index 1 is Showing as Outlier which we Assign Extreme value above

```python
#Anomaly Scores
data_encoded['scores']=clf.decision_function(data_encoded)
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:10])
data_encoded
```

Inference: Anomaly Score -1 are Outliers so 1th record is an Outlier

```python
#Print the outlier data points
data_encoded[data_encoded['anomaly']==-1]
```

Inference: Checking Anomaly Score -1 are Outliers
